# Pr√©dction en direct

```python
# === Imports n√©cessaires ===
from IPython.display import display, Javascript
from google.colab.output import eval_js
import cv2
import numpy as np
import base64
import time
from ultralytics import YOLO

# === Charger ton mod√®le YOLOv8 ===
model = YOLO("/content/drive/MyDrive/yolov8_best.pt")

# === Afficher une image vide pour d√©marrer ===
display(Javascript('''
  if (!document.getElementById("live_img")) {
    const img = document.createElement('img');
    img.id = "live_img";
    img.style.width = "640px";
    document.body.appendChild(img);
  }
'''))

# === JavaScript modifi√© : capture avec bouton stop ===
js_code = """
var stopCapture = false;

async function captureWebcam() {
  const video = document.createElement('video');
  video.width = 640;
  video.height = 480;
  video.autoplay = true;
  video.muted = true;

  const stream = await navigator.mediaDevices.getUserMedia({video: true});
  video.srcObject = stream;
  await video.play();

  const canvas = document.createElement('canvas');
  canvas.width = video.videoWidth;
  canvas.height = video.videoHeight;
  const context = canvas.getContext('2d');

  if (!document.getElementById('stopButton')) {
    const stopButton = document.createElement('button');
    stopButton.id = 'stopButton';
    stopButton.innerText = 'Stop Capture';
    stopButton.style.marginTop = '10px';
    stopButton.onclick = () => { stopCapture = true; };
    document.body.appendChild(stopButton);
  }

  async function capture() {
    if (stopCapture) {
      stream.getTracks().forEach(track => track.stop());
      return;
    }
    context.drawImage(video, 0, 0, canvas.width, canvas.height);
    const data = canvas.toDataURL('image/jpeg', 0.8);
    await google.colab.kernel.invokeFunction('notebook.capture', [data], {});
    setTimeout(capture, 300);
  }
  capture();
}
"""

# Execute the JavaScript code to define captureWebcam function
display(Javascript(js_code)) # This line is added to define captureWebcam in JavaScript context


def capture():
    eval_js('captureWebcam()')

# === Fonction Python pour traiter chaque image captur√©e ===
def process_image(image_data):
    try:
        image_bytes = image_data.split(',')[1]
        decoded = np.frombuffer(base64.b64decode(image_bytes), dtype='uint8')
        frame = cv2.imdecode(decoded, cv2.IMREAD_COLOR)

        if frame is not None:
            # YOLOv8 pr√©diction
            results = model.predict(source=frame, conf=0.3, save=False)

            # Annoter l'image
            annotated_frame = results[0].plot()

            # Convertir en base64
            _, buffer = cv2.imencode('.jpg', annotated_frame)
            jpg_as_text = base64.b64encode(buffer).decode()

            # Mettre √† jour l'image sans clear_output
            display(Javascript(f'''
              document.getElementById("live_img").src = "data:image/jpeg;base64,{jpg_as_text}";
            '''))
            time.sleep(0.3)

    except Exception as e:
        print(f"Erreur pendant le traitement: {e}")

# === Connexion Python ‚ÜîÔ∏è JavaScript ===
from google.colab import output
output.register_callback('notebook.capture', process_image)

# === Lancer la capture
capture()
```

```python
%%writefile app.py
import streamlit as st
from ultralytics import YOLO
from PIL import Image
import tempfile
import os
import shutil

st.set_page_config(page_title="Trash Detection App", layout="centered")
st.title("üóëÔ∏è D√©tection de d√©chets avec YOLOv8")
st.write("Chargez une **image** et le mod√®le d√©tectera les d√©chets.")

@st.cache_resource
def load_model():
    return YOLO("/content/drive/MyDrive/yolov8_best.pt")

model = load_model()

if os.path.exists("runs/detect/predict"):
    shutil.rmtree("runs/detect/predict")

uploaded_file = st.file_uploader("üì∑ Choisissez une image :", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="üñºÔ∏è Image originale", use_column_width=True)
    with tempfile.NamedTemporaryFile(delete=False, suffix=".jpg") as temp_file:
        image.save(temp_file.name)
        model.predict(source=temp_file.name, save=True)
        result_dir = "runs/detect/predict"
        result_files = os.listdir(result_dir)
        if result_files:
            result_path = os.path.join(result_dir, result_files[0])
            st.image(result_path, caption="‚úÖ R√©sultat de la d√©tection", use_column_width=True)
        else:
            st.error("Erreur : aucune image r√©sultat trouv√©e.")
```
